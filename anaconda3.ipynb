5주차 실습문제 예제 3.1

리스트에 n개의 항목들이 들어있는데, 이들을 키의 순서에 따라 오름차순으로 재배치하라.

정렬되지 않은 리스트A가 입력되면 리스트의 길이를 n에 저장하고 첫번째 for문0부터 n-1가지 반복하면서 i값을 least변수에 대입, 두번째 for문으로 가서 리스트의 i번째 요소와 j번째요소를 비교해 j번째 요소가 더 작을때 까지 j만큼 반복, 그 중에 j번째가 더 작으면 그 요소를 least에 넣고 i번째 요소와 j번째 요소를 서로 교환하는 것을 반복 교환할때마다 중간과정 출력문장을 계속 출력

def selection_sort(A): #함수 선언 n = 리스트의 길이 for i in range(n-1): #0 ~n-2반복 for j in range(i + 1,n): # i+1 ~ n-1 반복 if A[j] < A[least]:\ # 비교 연산 least = j # 최소항목 갱신 A[i], A[least] = A[least],A[i] # 배열 항목 교환 printStep(A, i + 1) # 중간 과정 출력용 문장

```python
def selection_sort(A) :
   n = len(A)
   for i in range(n-1) :
      least = i
      for j in range(i+1, n) :
         if (A[j]<A[least]) :
            least = j
   A[i], A[least] = A[least], A[i]
   printStep(A, i + 1)
```
