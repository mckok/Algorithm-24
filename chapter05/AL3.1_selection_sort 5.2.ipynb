정렬된 두 리스트의 병합

병합 알고리즘 merge() 을 사용하여 병합된 리스트를 저장하기 위해 전역변수로 선언된 임시 리스트 sorted를 사용하였다.

1행 A[left.right]를 오름차순으로 정렬  
2행 항목이 2개 이상인 경우  
3행 리스트의 균등 분할  
4행 부분 리스트 정렬  
5행 부분 리스트 정렬  
6행 병합

```python
def merge(A, left, mid, right) :
    k = left
    i = left
    j = mid + 1
    while i<=mid and j<=right :
      if A[i] <= A[j] :
         sorted[k] = A[i]
         i, k = i+1, k+1
      else:
          sorted[k] = A[j]
          j, k = j+1, k+1
    
    if i > mid :
        sorted[k:k+right-j+1] = A[j:right+1]
    else :
        sorted[k:k+mid-i+1] = A[i:mid+1]
        
    A[left:right+1] = sorted[left:right+1]
```

```python
def merge_sort(A, left, right) :
    if left<right :
        mid = (left + right) // 2
        merge_sort(A, left, mid)
        merge_sort(A, mid + 1, right)
        merge(A, left, mid, right)
        
def merge(A, left, mid, right) :
    k = left
    i = left
    j = mid + 1
    while i<=mid and j<=right :
      if A[i] <= A[j] :
         sorted[k] = A[i]
         i, k = i+1, k+1
      else:
          sorted[k] = A[j]
          j, k = j+1, k+1
    
    if i > mid :
        sorted[k:k+right-j+1] = A[j:right+1]
    else :
        sorted[k:k+mid-i+1] = A[i:mid+1]
        
    A[left:right+1] = sorted[left:right+1]
    
data = [ 5, 3, 8, 4, 9, 1, 6, 2, 7]
sorted = [0] * len(data)
print("Original : ", data)
merge_sort(data, 0, len(data)-1)
print("MergeSort :", data)
```
